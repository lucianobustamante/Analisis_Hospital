```python
# ahora vamos a ver si hay valores faltantes
wine.isnull().sum()
```

```python
# ya que son pocos, podemos quitar estas lineas
wine = wine[wine['PESO_BEBE'].notna()]
wine = wine[wine['SEMANA_GESTACION'].notna()]
wine = wine[wine['PESO_GANADO'].notna()]
wine = wine[wine['PESO_DESPUES_EMB'].notna()]
wine = wine[wine['PESO_ANTES_EMB'].notna()]
```

```python
# comprobamos ahora que se hayan eliminado
print(wine['PESO_BEBE'].isnull().sum())
print(wine['SEMANA_GESTACION'].isnull().sum())
print(wine['PESO_GANADO'].isnull().sum())
print(wine['PESO_DESPUES_EMB'].isnull().sum())
print(wine['PESO_ANTES_EMB'].isnull().sum())
```

```python
sns.boxplot(y = 'PESO_GANADO', data = wine)
#LA MAYORIA DE LAS MUJERES GANO CASI 30 KILOS, PARA SER EXACTO 26,56 EN PROMEDIO 
```

```python
# Primero empezamos con el scatter plot
sns.relplot(x="PESO_BEBE", y="EDAD_MADRE", data=wine)
#podemos observar, que por ejemplo una madre de 45 a√±os vario significativamente el peso del bebe pasando los 7 kilos
# otro caso que podemos apreciar, es que en los adolecentes el peso tiende a ser mas bajo 
```
